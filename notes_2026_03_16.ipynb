```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

## `odeint`

```python
# mx'' + cx' + kx = 0 , x(0) = x0 , x'(0) = 0
# u0 = x , u1 = x' , U = [u0 u1]
m = 1; c = 1; k = 10; tf = 7; x0 = 4;
f = lambda u,t: [u[1] , (-k*u[0] - c*u[1])/m]
U0 = [x0,0]
t = np.linspace(0,tf,501)
U = spi.odeint(f,U0,t)
```
