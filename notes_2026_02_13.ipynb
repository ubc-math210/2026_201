## Vectorization

Vectorization of a function/operation is the process of applying it to every entry in a vector. Consider a vector
$$
\mathbf{x} = \begin{bmatrix} x_0 & x_1 & \cdots & x_N \end{bmatrix}
$$
Vector addition and scalar multiplication are vectorized:
$$
\mathbf{x} + \mathbf{y} = \begin{bmatrix} x_0 + y_0 & x_1 + y_1 & \cdots & x_N + y_N \end{bmatrix}
$$
$$
c \mathbf{x} = \begin{bmatrix} cx_0 & cx_1 & \cdots & cx_N \end{bmatrix}
$$
Consider other operations:
$$
\mathbf{x}^2 = \begin{bmatrix} x_0^2 & x_1^2 & \cdots & x_N^2 \end{bmatrix}
$$
The vectorization of a function $f(x)$ is
$$
f(\mathbf{x}) = \begin{bmatrix} f(x_0) & f(x_1) & \cdots & f(x_N) \end{bmatrix}
$$

## Plotting Examples

The plotting process:

1. Define vector of $x$ values
2. Use vectorization to compute vector of $y$ values
3. Use `plt.plot`
4. Apply style and formattting
5. Call `plt.show` to print the figure

Plot the function $y = e^{-x^2} \cos(x)$ on $[-2,2]$.

```python
x = np.linspace(-2,2,101)
y = np.exp(-x**2)*np.cos(5*x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-1,1,5)
y = np.sin(np.pi*x)
plt.plot(x,y)
plt.show()
```

How do we plot a parameteric curve $(x(t),y(t))$?

```python
t = np.linspace(0,2*np.pi)
x = np.cos(t)
y = np.sin(t)
plt.plot(x,y), plt.axis('equal')
plt.show()
```

```python
t = np.linspace(0,2*np.pi,200)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.plot(x,y,'r')
plt.show()
```
