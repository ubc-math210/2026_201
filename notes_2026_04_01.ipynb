```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
```

## Example 1

```python
x = np.array([0.,np.pi/2,np.pi])
y = np.array([1.,2.,1.5])
A = np.column_stack([np.ones(3),np.sin(x),np.cos(x)])
print(A)
```

```python
c = la.solve(A,y)
print(c)
```

```python
xs = np.linspace(-np.pi/2,3*np.pi/2,100)
ys = c[0] + c[1]*np.sin(xs) + c[2]*np.cos(xs)
plt.plot(xs,ys,x,y,'r.')
plt.show()
```

## Example 2

```python
# Make some fake data
# y = a + bx
a = -2.4; b = 5.1;
N = 50
x = np.random.rand(N+1) # N+1 random numbers between 0 and 1 uniformly distributed
error = 0.1*np.random.randn(N+1) # N+1 small random numbers with mean 0
y = a + b*x + error
plt.plot(x,y,'.')
plt.show()
```

```python
# Fit a linear model: f(x) = c0 + c1*x
A = np.column_stack([np.ones(N+1), x])
c = la.solve(A.T@A,A.T@y)
xs = np.linspace(0,1)
ys = c[0] + c[1]*xs
plt.plot(xs,ys,x,y,'r.')
plt.show()
```

## Example 3

```python
# Make some fake data
# y = a0 + a1 x + a2 x^2
a0 = -2.4; a1 = 5.1; a2 = -1.1;
N = 100
x = 5*np.random.rand(N+1) # N+1 random numbers between 0 and 5 uniformly distributed
error = 0.5*np.random.randn(N+1)
y = a0 + a1*x + a2*x**2 + error

# Fit a quadratic model: f(x) = c0 + c1*x + c2*x**2
A = np.column_stack([np.ones(N+1), x, x**2])
c = la.solve(A.T@A,A.T@y)
xs = np.linspace(0,5,100)
ys = c[0] + c[1]*xs + c[2]*xs**2
plt.plot(xs,ys)
plt.plot(x,y,'r.')
plt.show()
```
