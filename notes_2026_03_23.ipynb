## Order of Accuracy

```python
def odeRK2(f,y0,t):
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(len(t) - 1):
        alpha = 3/4
        h = t[n + 1] - t[n]
        k1 = f(y[n],t[n])
        k2 = f(y[n] + alpha*k1*h,t[n] + alpha*h)
        y[n + 1] = y[n] + (k1 + 2*k2)*h/3
    return y
```

```python
f = lambda y,t: -y
y0 = 1
t = np.linspace(0,2,100)
ya = odeRK2(f,y0,t)
yb = spi.odeint(f,y0,t)
plt.plot(t,ya,'.',t,yb)
plt.show()
```

```python
f = lambda y,t: np.cos(y) - y
y0 = 1
t = np.linspace(0,2,100)
ya = odeRK2(f,y0,t)
yb = spi.odeint(f,y0,t)
plt.plot(t,ya,'.',t,yb)
plt.show()
```

## Solving Linear Systems

We want to solve linear systems $A \mathbf{x} = \mathbf{b}$.

```python
N = 100
A = -2*np.eye(N) + np.diag(np.ones(N-1),1) + np.diag(np.ones(N-1),-1)
b = np.ones((N,1))
x = la.solve(A,b)
```
