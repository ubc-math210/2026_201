## Example 1

Find all roots of $p(x) = x^3 + x - 1$. We proved that there is exactly 1 root and it is in the interval $(0,1)$.

Translate the equation $p(x) = 0$ into $x = f(x)$. Try $f(x) = 1 - x^3$.

```python
x = 1.5; N = 5;
f1 = lambda x: 1 - x**3
for n in range(1,N+1):
    x = f1(x)
    print(x)
```

Try $f(x) = (1 - x)^{1/3}$.

```python
x = 0.5; N = 1000;
f2 = lambda x: (1 - x)**(1/3)
for n in range(1,N+1):
    x = f2(x)
print(x)
```

```python
f2(0.6823278038280193)
```

Try $f(x) = \frac{1}{x^2 + 1}$.

```python
x = 0.5; N = 50;
f3 = lambda x: 1/(x**2 + 1)
for n in range(1,N+1):
    x = f3(x)
print(x)
```
