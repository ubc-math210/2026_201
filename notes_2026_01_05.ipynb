## Cells and Modes

There are 2 types of cells: markdown and code. We write text, markdown, HTML, and LaTeX in markdown cells. We write Python code in code cells.

There are 2 kinds of modes: edit and command. We are in edit mode when we see the cursor in the cell. And we are in command mode when we don't see the cursor and the cell is outlined in blue. Double click a cell (or presss `ENTER`) to enter edit mode. And click `ESC` to enter command mode.

## LaTeX

LaTeX is a development environment for creating scientific documents. We are going to use just the part of LaTeX for rendering mathematical notation.

Write LaTeX inline with single dollar signs `$ ... $`.

For example, we favourite function is $f(x) = e^{-x^2} \cos(x)$.

Write LaTeX in display mode using double dollar signs `$$ ... $$`.

For example, the Fundamental Theorem of Calculus is:

$$
\int_a^b f(x) dx = F(b) - F(a)
$$
