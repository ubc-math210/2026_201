## Riemann Sums

Approximate the following intergral using left and right Riemann sums (with $N+1$ evenly spaced points) and compare with the exact value:

$$
\int_0^1 e^{-x} dx = 1 - \frac{1}{e}
$$

First plot the function:

```python
x = np.linspace(0,1,50)
y = np.exp(-x)
plt.plot(x,y), plt.ylim([-0.25,1.25]), plt.grid(True)
plt.show()
```

```python
a = 0; b = 1; N = 100; dx = (b - a)/N;
x = np.linspace(a,b,N + 1)
xL = x[:-1]
yL = np.exp(-xL)
L = np.sum(yL)*dx
print(L)
```

```python
a = 0; b = 1; N = 100; dx = (b - a)/N;
x = np.linspace(a,b,N + 1)
xR = x[1:]
yR = np.exp(-xR)
R = np.sum(yR)*dx
print(R)
```
