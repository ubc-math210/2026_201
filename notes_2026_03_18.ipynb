```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

```python
m = 1; k = 2; L = 1;
def f(u,t):
    du0dt = u[1]
    du1dt = k/m*(L*u[0]/(u[0]**2 + u[2]**2)**0.5 - u[0])
    du2dt = u[3]
    du3dt = k/m*(L*u[2]/(u[0]**2 + u[2]**2)**0.5 - u[2])
    dudt = [du0dt,du1dt,du2dt,du3dt]
    return dudt

U0 = [0,-1,1,0]
t = np.linspace(0,100,2001)
U = spi.odeint(f,U0,t)
plt.plot(U[:,0],U[:,2])
plt.show()
```
