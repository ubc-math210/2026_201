## Secant Method

Write a function called `secant` which takes function `f`, numbers `a` and `b`, integer `N`, and return the $x$-intercept os secant line of $N$th interval from secant method.

```python
def bisection(f,a,b,N):
    an = a
    bn = b
    for n in range(1,N+1):
        mn = (an + bn)/2
        if f(an)*f(mn) < 0:
            bn = mn
        elif f(mn)*f(bn) < 0:
            an = mn
        elif f(mn) == 0:
            print("Found exact value.")
            return mn
        else:
            print("Bisection fails.")
    return (an + bn)/2
```

```python
def secant(f,a,b,N):
    an = a
    bn = b
    for n in range(1,N+1):
        mn = an - f(an)*(bn - an)/(f(bn) - f(an))
        if f(an)*f(mn) < 0:
            bn = mn
        elif f(mn)*f(bn) < 0:
            an = mn
        elif f(mn) == 0:
            print("Found exact value.")
            return mn
        else:
            print("Secant method fails.")
    return an - f(an)*(bn - an)/(f(bn) - f(an))
```

```python
f = lambda x: x**2 - 2
a = 1
b = 2
N = 5
r1 = secant(f,a,b,N)
r2 = bisection(f,a,b,N)
r = 2**0.5
print("Secant:",r1,"Error:",abs(r - r1))
print("Bisection:",r2,"Error:",abs(r - r2))
```

```python
f = lambda x: x**100 - 0.1
a = 0
b = 1
N = 10
r1 = secant(f,a,b,N)
r2 = bisection(f,a,b,N)
r = 0.1**(1/100)
print("Secant:",r1,"Error:",abs(r - r1))
print("Bisection:",r2,"Error:",abs(r - r2))
```

## Newton's Method

Write a function called `newton` which takes functions `f` and `df`, numbers `x0` and `N`, and returns the value after $N$ iterations of Newton's method applied to $f(x)$ with initial value $x_0$.

```python
def newton(f,df,x0,N):
    xn = x0
    for n in range(N):
        xn = xn - f(xn)/df(xn)
    return xn
```

```python
f = lambda x: x**2 - 2
df = lambda x: 2*x
x0 = 1
N = 4
r1 = newton(f,df,x0,N)
r = 2**0.5
print("Newton:",r1,"Error:",abs(r - r1))
```
