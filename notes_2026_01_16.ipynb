## `def` statement

Construct our own function using the `def` statement:

```
def fun(x,y,z):
    "Documentation string to explain the function."
    Python code
    Python code
    Python code
    return value
```

where:

* keyword `def` begins the definition
* `fun` is the function name
* `x,y,z` are input parameters
* colon `:` ends the `def` statement
* documentation string on second line is optional
* body of the function indented 4 spaces
* `return` keyword specifies the output value

Write a function called `average` which takes a Python list of numbers `x` and returns the average value.

Let's develop a systematic approach to writing functions:

1. Write a stub. That is the minimum amount of code to return a value of the correct type.
2. Come up with simple examples of input values where we know the exact value of the output.
3. Write code and test often!

````python
# Write a stub
def average(x):
    return 0.0
````

````python
def average(x):
    "Compute the average value of entries in sequence x."
    total = sum(x)
    length = len(x)
    return total/length
````

````python
x1 = [1,2,3]
result1 = average(x1)
print(result1,2.0)
````

````python
x2 = [-3,-1,2,4]
result2 = average(x2)
print(result2,0.5)
````

## `lambda` construction

Define simple functions in a single line using the `lambda` keyword.

````python
average = lambda x: sum(x)/len(x)
````

````python
average([1,2,3,4,5,6])
````
