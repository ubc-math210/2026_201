Approximate the following integral with a left Riemann sum and find a bound for the error: 
$$
\int_0^{\sqrt{\pi/2}} \sin(x^2) dx
$$

```python
a = 0; b = np.sqrt(np.pi/2); N = 1969;
dx = (b - a)/N
f = lambda x: np.sin(x**2)
x = np.linspace(a,b,N+1)
y = f(x)
L = dx*np.sum(y[:-1])
K1 = np.sqrt(2*np.pi)
E = (b - a)**2/2/N*K1
print(L,"+-",E)
```
