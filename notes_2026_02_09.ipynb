## Python Modules and Packages

A Python module is a text file (with extension `.py`) contatining Python code. A Python package is directory of files containing code. We import modules/packages with the `import` keyword and access data, function, objects, etc. inside the module/package using the dot syntax. 

Let's create a module called `roots` which contains all our root finding methods.

````python
import roots
````

````python
f = lambda x: x**2 - 2
a = 1
b = 2
N = 10
r = roots.bisection(f,a,b,N)
print(r)
````

````python
f = lambda x: x**2 - 2
Df = lambda x: 2*x
x0 = 1.5
eps = 1e-10
N = 5
r = roots.newton(f,Df,x0,eps,N)
print(r)
````

Plot the graph $y = e^{-x^2} \cos(10x)$ on $[0,2]$.

Let $f(x)$ be a function and let $\mathbf{x}$ be a vector
$$
\mathbf{x} = \begin{bmatrix} x_0 & x_1 & \cdots & x_N \end{bmatrix}
$$
Then $f(\mathbf{x})$ is the vector
$$
f(\mathbf{x}) = \begin{bmatrix} f(x_0) & f(x_1) & \cdots & f(x_N) \end{bmatrix}
$$
This is the **vectorization** of $f(x)$.

Mathematical functions in Python are vectorized. 

In particular, the expression above
```
y = np.exp(-x**2)*np.cos(10*x)
```
generates the vector
$$
\mathbf{y} = \begin{bmatrix} e^{-x_0^2}\cos(10x_0) & e^{-x_1^2}\cos(10x_1) & \cdots & e^{-x_N^2}\cos(10x_N) \end{bmatrix}
$$

````python
x = np.linspace(0,2,201)
y = np.exp(-x**2)*np.cos(10*x)
plt.plot(x,y)
plt.show()
````
