## Solve Systems of Equations

```python
N = 4
A = -2*np.eye(N) + np.diag(np.ones(N-1),1) + np.diag(np.ones(N-1),-1)
A
```

```python
b = np.array([1.,1.,2.,1.])
```

```python
x = la.solve(A,b)
x
```

```python
b1 = np.array([[1.],[1.],[2.],[1.]])
b1
```

```python
b1 = np.array([1.,1.,2.,1.]).reshape(4,1)
b1
```

```python
x = la.solve(A,b1)
x
```

## Interpolation

```python
x = np.array([-1.,0.,1.])
y = np.array([0.,1.,0.])
A = np.vander(x,increasing=True)
#A = np.column_stack([x**0,x**1,x**2])
c = la.solve(A,y)
c
```

```python
d = 10
x = np.arange(d+1)
y = np.random.randint(0,10,d+1)
A = np.vander(x,increasing=True)
c = la.solve(A,y)
xs = np.linspace(0,d,200)
ys = sum([c[n]*xs**n for n in range(d+1)])
plt.plot(x,y,'.',xs,ys)
plt.show()
```
