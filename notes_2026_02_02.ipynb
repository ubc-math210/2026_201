## Bisection Method

Write a function called `bisection` which takes a function `g`, numbers `a` and `b`, and integer `N`, and returns the midpoint after `N` iterations of bisection method.

```python
def bisection(g,a,b,N):
    an = a
    bn = b
    for n in range(1,N+1):
        mn = (an + bn)/2
        if g(an)*g(mn) < 0:
            bn = mn
        elif g(mn)*g(bn) < 0:
            an = mn
        elif g(mn) == 0:
            print("Found exact value.")
            return mn
        else:
            print("Bisection fails.")
    return (an + bn)/2
```

```python
g = lambda x: x**2 - 2
a = 1; b = 2; N = 2;
r = bisection(g,a,b,N)
print(r)
```

## Example

Find all fixed points of $f(x) = \sqrt{1 + x} + \sqrt{x}$.

```python
g = lambda x: x - (1 + x)**0.5 - x**0.5
a = 4
b = 5
N = 30
r = bisection(g,a,b,N)
print(r)
```
