## Example

Write a function called `poly_root` which takes a list of numbers `p` (representing a polynomial $p(x)$), number `x0` and positive integer `N` (default value $N=10$), and return $x_N$ the value after $N$ iterations of Newton's method applied to $p(x)$ with initial value $x_0$.

```python
def newton(f,df,x0,N):
    xn = x0
    for n in range(N):
        xn = xn - f(xn)/df(xn)
    return xn

def poly_eval(p,a):
    terms = [p[n]*a**n for n in range(len(p))]
    return sum(terms)

def poly_diff(p):
    if len(p) == 1:
        return [0]
    else:
        dp = [n*p[n] for n in range(1,len(p))]
        return dp

def poly_root(p,x0,N=10):
    f = lambda x: poly_eval(p,x)
    df = lambda x: poly_eval(poly_diff(p),x)
    return newton(f,df,x0,N)
```

```python
p = [-2,0,1]
x0 = -1
r = poly_root(p,x0)
print(r,-2**0.5)
```

```python
p = [-5,1,0,1,0,1]
x0 = 1.5
r = poly_root(p,x0)
print(r)
```
