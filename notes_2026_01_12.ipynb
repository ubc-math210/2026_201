A range object represents a sequence of evenly-spaced integers. Create a range object with `range(a,b,step)` which is integers from `a` to `b` (exclusive) incremented by `step`.

The function `range` doesn't create the values but instead the range object returns the integers in the sequence when we ask for it such as in a `for` loop.

Use a list comprehension and the function `sum` to compute
$$
\sum_{n=1}^{100} \frac{1}{n^2}
$$

```python
terms = [1/n**2 for n in range(1,11)]
```

```python
print(terms)
```

```python
sum(terms)
```

Write a list comprehension to create the sequence of values from 0 to 1 (inclusive) incremented by 0.1.

```python
[n/10 for n in range(0,11)]
```

Compute the Riemann sum
$$
\sum_{k=1}^N \frac{1}{1 + x_k^2}
$$
where $x_k = k/N$ for $k=1,\dots,N$.

```python
N = 10000
I = 1/N*sum([1/(1 + (k/N)**2) for k in range(1,N+1)])
print(I)
```

This is an approximation of the definite integral
$$
\int_0^1 \frac{dx}{1 + x^2} = \frac{\pi}{4}
$$

```python
3.1415926/4
```
