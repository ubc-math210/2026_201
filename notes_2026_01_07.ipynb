## LaTeX

### Inline and Display

Write LaTeX code within single dollar signs `$ ... $` to display mathematical notation inline. For example, we have the trigonometric identity $\sin^2 \theta + \cos^2 \theta = 1$.

Write LaTeX code within double dollar signs `$$ ... $$` to display mathematical notation in its own centered paragraph. For example, the definition of the derivative of $f(x)$ at $x = a$ is

$$
f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}
$$

### Examples

The mean value theorem states that for a differentiable function $f(x)$ there exists $c \in (a,b)$ such that

$$
f'(c) = \frac{f(b) - f(a)}{b - a}
$$

The antiderivative of $f(x) = \frac{1}{1 + x^2}$ is

$$
\int \frac{dx}{1 + x^2} = \arctan(x) + C
$$

The Taylor series of $e^x$ is

$$
e^x = \sum_{k = 0}^{\infty} \frac{x^k}{k!}
$$

Formula 3.363.2 on page 345 of the Table of Integrals, Series and Products is

$$
\int_u^{\infty} \frac{e^{-\mu x} dx}{x \sqrt{x - u}}  = \frac{\pi}{\sqrt{u}} [1 - \Phi(\sqrt{u \mu})]
$$
